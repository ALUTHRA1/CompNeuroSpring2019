# Python, NumPy and Matplotlib basic examples

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d  # for 3d plots
```

### Create a 5 x 10 array of zeros.

```python
np.zeros((5, 10))
```

### Create a 10 x 2 array of random numbers between 0 and 1 and assign it to a variable.

```python
xy = np.random.rand(10, 2)
print(xy.shape)
print(xy.dtype)
xy
```

### Treat each row of the above array as an (x, y) coordinate pair. Create two views into the array named x and y.

```python
x = xy[:,0]
y = xy[:,1]
x
```

### Verify that x and y are views into the same data as the original array by setting the first (x, y) pair to (0, 1).

```python
x[0] = 0
y[0] = 1
xy
```

### Add 1 to every element of the entire array.

```python
xy += 1
xy
```

### Make a length 10 array of integers from 1 to 10. Notice that the index array is one dimensional, such that whether it is a column or row vector is indeterminant. 

```python
indexes = np.arange(1, 11)
print(indexes.shape)
print(indexes.dtype)
indexes
```

### Reshape the index array into a column vector.

```python
indexes = indexes.reshape((10, 1))
print(indexes.shape)
indexes
```

### Multiply each column of x and y values by the index array.

```python
xy *= indexes  # both columns will each be multiplied by indexes via broadcasting
xy
```

### Store the min/max range for both x and y coordinates as tuples.

```python
xlims = (x.min(), x.max())  # x and y are still views into the same memory as arr
ylims = (y.min(), y.max())
xlims, ylims
```

### Print the maximum y value from the above tuples.

```python
print("y-max =", ylims[1])
```

### Create a new array with distances from the origin of each (x, y) pair.

```python
xy**2
```

```python
distances = np.sqrt((xy**2).sum(axis=1))
distances
```

### What are the mean x and y coordinates for all points?

```python
xy.mean(axis=0)
```

### Assign the mean x and y coords each to their own variable.

```python
xavg, yavg = xy.mean(axis=0)
xavg, yavg
```

### Subtract the mean (xavg,yavg) position from each (x,y) coordinate pair.

```python
xy -= np.array([xavg, yavg])
xy
```

```python
# Note that x and y still point to the columns of xy.
x, y
```

```python
xy = xy + 10  # this assigns xy to a new set of values independent of the previous column views x and y
xy, x, y
```

### Reflect the shifted (x,y) pairs about the y-axis.

```python
xy[:,0] = -xy[:,0]
xy
```

### Add 2 to each x value with absolute value less than 3.

```python
print(xy.shape)
num_rows = xy.shape[0]
for row in range(num_rows):
    if abs(xy[row, 0]) < 3:
        xy[row, 0] += 2
xy
```

### Subtract 2 from each y value whose (x,y) point is a distance of more than 3 from the origin.

```python
distances = np.sqrt((xy**2).sum(axis=1))
for row, distance in enumerate(distances):
    if distance > 3:
        xy[row, 1] -= 2
xy
```

### Plot a line through the (x,y) points.

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

### Plot the (x,y) points as a scatter plot.

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y');
```

### Overlay scatter plots of the points before and after shifting by (+1,+1).

```python
plt.scatter(x, y, label='original')
plt.scatter(x + 1, y + 1, label='shifted')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

### Repeat above overlay and connect points by dashed lines.

```python
plt.scatter(x, y, label='original')
plt.plot(x, y, '--')
plt.scatter(x + 1, y + 1, label='shifted')
plt.plot(x + 1, y + 1, '--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

### Plot the original and shifted (x,y) pairs side-by-side with different marker symbols and different colors.

```python
fig = plt.figure(figsize=(10, 5))

ax = fig.add_subplot(1, 2, 1)  # !!! <== subplot index starts at 1, NOT 0 !!!
ax.scatter(x, y, marker='o', facecolor='none', edgecolor='blue', s=500)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('original')

ax = fig.add_subplot(1, 2, 2)
ax.scatter(x + 1, y + 1, marker='s', facecolor='none', edgecolor='red', s=100)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('shifted');
```

### Create 6 sets of 10 random (x,y) pairs and plot them separately in a 2 x 3 grid.

```python
xy = np.random.rand(6, 10, 2);  # 6 x 10 x 2
print(xy.shape)
print(xy[0, :, :])  # first set

fig = plt.figure()
for i in range(6):
    ax = fig.add_subplot(2, 3, i+1)  # !!! <== subplot index starts at 1, NOT 0 !!!
    ax.scatter(xy[i, :, 0], xy[i, :, 1])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
fig.tight_layout();  # helps improve margins between plots
```

### Create 100 (x, y, z) random points within a unit cube and plot them in 3D.

```python
from mpl_toolkits import mplot3d  # for 3d plots

xyz = np.random.rand(100, 3)  # 100 x 3, each row is and x, y, z coordinate
x = xyz[:, 0]
y = xyz[:, 1]
z = xyz[:, 2]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.scatter(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
fig.tight_layout();  # helps improve margins between plots
```

### Repeat the above and add a dashed line from the x-y plane to each point.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.scatter(x, y, z)
ax.scatter(x, y, 0, marker='.', color='red', alpha=0.25)
for i in range(xyz.shape[0]):
    ax.plot([x[i], x[i]], [y[i], y[i]], [0, z[i]], '--', color='black', alpha=0.25)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
fig.tight_layout();  # helps improve margins between plots
```

### Create a 10 x 10 matrix of random values between 0 and 30 and plot it as a colormap.

```python
mat = np.random.rand(10, 10) * 30

plt.imshow(mat)
plt.xlabel('column')
plt.ylabel('row')
plt.colorbar()

mat
```

### Clamp colorscale for above matrix to 10-20.

```python
plt.imshow(mat, vmin=10, vmax=20)
plt.xlabel('column')
plt.ylabel('row')
plt.colorbar();
```

### Repeat with custom colormap.

```python
import matplotlib as mpl

plt.imshow(mat, cmap=mpl.cm.nipy_spectral_r)
plt.xlabel('column')
plt.ylabel('row')
plt.colorbar();
```
