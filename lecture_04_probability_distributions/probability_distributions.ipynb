# Exponential distribution example plotting and maximum likelihood fitting

```python
# Exponential distribution
rate = 1
sample_data = np.random.exponential(1 / rate, (1000,))  # random samples from exponential distribution with given rate
```

```python
fig = plt.figure()
nbins = 50
counts, bins, patches = plt.hist(sample_data, nbins, density=True, label='data', alpha=0.5)
plt.xlabel('data values')
plt.ylabel('PDF')

# Maximum likelihood fit to sample data
# loc: shifted start of exponential (here we'll fix it to be zero)
# scale: 1 / rate
loc, scale = st.expon.fit(sample_data, floc=0)
rate_estimate = 1 / scale
my_pdf = st.expon.pdf(bins, loc=0, scale=scale)
plt.plot(bins, my_pdf, label='pdf')
plt.legend()

loc, rate_estimate
```
